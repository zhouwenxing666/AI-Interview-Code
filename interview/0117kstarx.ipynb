# 四则运算 实现eval

```python
# 不带括号()
# 以 1+12*5-9 为例：
# 遇到 +/-：把 num 直接（带符号）压栈

# 遇到 * /：立刻把栈顶数字弹出，与当前 num 计算后再压回

# 最终栈里只剩下需要做加减的项，直接 sum(stack) 完成加减。

# 这是一种“把乘除优先级提前消化”的技巧，常见于题目 Basic Calculator II。
from collections import deque

def calculate_0202(s: str) -> int:
    #step1 定义一个栈 存数字
    deq = deque()
    
    #step2 "1+12*5-9"中的1可以看成是'+'1,前面有'+'号
    tmp_sign = '+'
    tmp_num = 0

    #step3 遍历s
    for i, ch in enumerate(s):

        #step3.1 如果当前字符ch是数字
        if ch.isdigit():
            tmp_num = tmp_num * 10 + int(ch)
        
        #step3.2 遇到运算符 或 到达末尾：结算 tmp_num
        if ch in '+-*/' or i == len(s) - 1:
            if tmp_sign == "+":
                deq.append(tmp_num)
            elif tmp_sign == "-":
                deq.append(-tmp_num)
            elif tmp_sign == "*":
                last_num = deq.pop()
                deq.append(last_num * tmp_num)
            else:# 除
                last_num = deq.pop()
                deq.append(last_num // tmp_num)
            
            #step3.3 更新tmp符号和num
            tmp_sign = ch
            tmp_num = 0
    
    return sum(deq)

print(calculate_0202("1+12*5-9"))  # 52



```
